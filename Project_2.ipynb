```python
df = df
df = df.replace(-200, np.nan)
df.isnull().sum()
```

```python
df['datetime'] = df['Date']  + pd.to_timedelta(df['Time'].astype(str))
df['datetime'] = pd.to_datetime(df['datetime'])
type(df['datetime'])
df.head()
```

```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Day'] = pd.DatetimeIndex(df['Date']).day
df['WeekDay'] = pd.DatetimeIndex(df['Date']).weekday
df.head()
```

```python
df = df.drop(columns = "NMHC(GT)")

df
```

```python
df = df.dropna(axis=0)

df
```

```python
df.rename(columns = {
    'CO(GT)' : 'CO in mg/m^3',
    'PT08.S1(CO)' : 'Tin Oxide(CO trgt)',
    'C6H6(GT)' : 'Benzene in microg/m^3',
    'PT08.S2(NMHC)' : 'Titania(NMHC trgt)',
    'NOx(GT)' : 'NOx in ppb',
    'PT08.S3(NOx)' : 'Tungsten Oxide(NOx trgt)',
    'NO2(GT)' : 'NO2 in microg/m^3',
    'PT08.S4(NO2)' : 'Tungsten Oxide(NO2 trgt)',
    'PT08.S5(O3)' : 'Indium Oxide(O3 trgt)',
    'T' : 'Temperature',
    'RH' : 'Relative Humidity',
    'AH' : 'Absolute Humidity',
    'datetime' : 'Datetime'}, inplace=True)

df
```

```python
df = df.drop(columns = [ "Time"])

df
```

```python
df["CO in mg/m^3"] *= 1000
df["NOx in ppb"] *= 1.9

df.rename(columns = {
    'CO in mg/m^3' : 'CO in ug/m^3',
    'Benzene in microg/m^3' : 'Benzene in ug/m^3',
    'NOx in ppb' : 'NOx in ug/m^3',
    'NO2 in microg/m^3' : 'NO2 in ug/m^3'}, inplace=True)
df
```

```python
df.to_csv('cleaned_air_quality_No_NA.csv')
```
